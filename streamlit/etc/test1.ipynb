### 일반 함수 vs `yield` 함수 비교
- 스트림릿 다 빼고 순수 파이썬으로 비교해보기

```python
# 일반 함수 (return)

def get_numbers() :
    return [1,2,3]

result = get_numbers()
print(result) # →  한 번에 전체가 튀어나옴
```

```python
# yield 함수 (제너레이터)
def get_numbers_slowly() :
    print('1을 준비중 ...')
    yield 1
    print('2를 준비중 ...')
    yield 2
    print('3을 준비중 ...')
    yield 3

gen = get_numbers_slowly()
print(gen) # <generator object ...>  → 아직 아무것도 실행 안 됨!

# gen = get_numbers_slowly()를 호출해도 함수 내부 코드가 실행되지 않는다는 점이에요. 
# "실행 준비"만 된 상태예요.
```

```python
print(next(gen))
# 1을 준비중...
# 1

print(next(gen))
# 2를 준비중...
# 2

print(next(gen))
# 3을 준비중...
# 3
```

```python
for num in get_numbers_slowly():
    print(f'받은 값 : {num}')
```

#### 핵심 한 줄 요약
- `return`은 "결과 전체를 한 번에" 주고, `yield`는 "값을 하나씩, 요청할 때마다" 준다.
